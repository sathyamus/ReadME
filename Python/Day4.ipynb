```python
total = 0
count = 0
while True :
    number = input('Enter a number: ')
    if number == 'done' : break
    value = float(number)
    total = total + value
    count += 1
average = total / count
print('Average:' , average)
print('Total:', total)
print('Count:', count)
```

```python
numlist = list()
while True :
    inp = input("Enter a numner")
    if inp == 'done' : break
    value = float(inp)
    numlist.append(value)
    
average = sum(numlist) / len(numlist)
print('Average:' , average)
print('Total:', sum(numlist))
print('Count:', len(numlist))
```
